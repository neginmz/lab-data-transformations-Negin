```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('files_for_lab/marketing_customer_analysis.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
#total number of responses in a plot

fig,x=plt.subplots()
x.bar(x = df['Response'].value_counts().index, height = df['Response'].value_counts().values)
plt.show()
```

```python
df.groupby(['Response','Sales Channel']
```

```python
#Show a plot of the response rate by the sales channel.
sns.histplot(binwidth=0.5, x="Response", hue="Sales Channel", data=df, stat="count", multiple="stack")
```

```python
#Show a plot of the response rate by the total claim amount.
fig, ax = plt.subplots()
sns.barplot(data=df, x="Response", y="Total Claim Amount", ax = ax)
plt.show()
```

```python
#Show a plot of the response rate by income.

fig, ax = plt.subplots()
sns.barplot(data=df, x="Response", y="Income")
```
